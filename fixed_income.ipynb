# Fixed Income

## Domain knowledge:
##### 1. Interest rate: Interest rate have inverse relationship with bond price, because when interest rate increase, the new bond will have higher yield leading to lower return from old bond
##### 2. Inflation: Inflation linked to bond price because if inflation increase the value of real bond will decline, leading to increase in demand for higher yield to offset inflation
##### 3. Credit risk; The risk that default on interest or principal payment that investors need to face, so higher credit risk will lead to higher yield to attract investors
##### 4. Liquidity: if liquidity is high, the yield need to compensate for the difficulty of buying bond
##### 5. Duration: the higher duration bond will be more sensitive with the change of the interest rate. For example: interest-5%, Future value = 100$; 1 year bond:  Present value = Future value/(1+0.05) = 95.24$, 10 years bond: Present value = Future value/(1+0.05)^10(due to compound interest) = 61,4$ -> the gap is significant between 1 year and 10 year: 5% and 40%
##### 6. Economic: weak economy tend to have higher demand on bond, pushing the bond price up because bond is a safe and stable, and the economy also cut interest to encourage growth, so the price for bond increase. Investors in strong economy will demand higher yield to offset inflation .
##### 7. Supply and Demand dynamic: The increase in government borrowing will lead to higher bond supply, so yield will increase and price fall, and investors will demand more bond in uncertain period such as recession.
##### 8. Currency risk( for foreign bond): the exchange rate movements may affect return, so high exchange rate will push the yield up to offset the currency depreciation, aiming to attract investors.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
uk = pd.read_csv('United Kingdom 10-Year Bond Yield Historical Data.csv', parse_dates= ['Date'])
china = pd.read_csv('China 10-Year Bond Yield Historical Data (1).csv', parse_dates= ['Date'])
usa = pd.read_csv('United States 10-Year Bond Yield Historical Data.csv', parse_dates= ['Date'])
japan = pd.read_csv('Japan 10-Year Bond Yield Historical Data.csv', parse_dates= ['Date'])
```

```python
uk.set_index('Date', inplace= True)
china.set_index('Date', inplace= True)
usa.set_index('Date', inplace= True)
japan.set_index('Date', inplace= True)
```

```python
uk = uk.sort_index().loc['2025-09-01':'2025-09-30']
china = china.sort_index().loc['2025-09-01':'2025-09-30']
usa = usa.sort_index().loc['2025-09-01':'2025-09-30']
japan = japan.sort_index().loc['2025-09-01':'2025-09-30']
```

# Month on Month analysis

```python
mom_uk = ((uk.iloc[-1]['Price'] - uk.iloc[0]['Price']) / uk.iloc[0]['Price'] * 100).round(2)
print('Month on Month percentage change of United Kingdom : ', mom_uk,'%')

mom_usa = ((usa.iloc[-1]['Price'] - usa.iloc[0]['Price']) / usa.iloc[0]['Price'] * 100).round(2)
print('Month on Month percentage change of United States : ', mom_usa,'%')

mom_china = ((china.iloc[-1]['Price'] - china.iloc[0]['Price']) / china.iloc[0]['Price'] * 100).round(2)
print('Month on Month percentage change of China : ', mom_china,'%')

mom_japan = ((japan.iloc[-1]['Price'] - japan.iloc[0]['Price']) / japan.iloc[0]['Price'] * 100).round(2)
print('Month on Month percentage change of Japan : ', mom_japan,'%')
```

##### Notes: From the data, we see that the yield rate in USA and UK turn negative at the end of the month, so this means the demand for bond increase, this signal the a weak growth or deflationary pressure of the economy. For China and Japan, the yield rate increase at the end of the month, signaling the strong growth in Japan and China, and investors intend to shift to riskier asset that can be grow such as equities(companies can gain more profits from higher purchasing power of customer), so they sell bonds making the price of them cheaper

```python
bond = pd.concat([uk['Price'], china['Price'], usa['Price'], japan['Price']], axis= 1)
```

# Countries bond rate analysis

```python
bond.columns = ['United Kingdom', 'China', 'United States','Japan']
bond = bond.loc['2025-09-01':'2025-09-30'].dropna()
bond
```

```python
import matplotlib.ticker as mtickers

bond['United Kingdom'].plot(figsize=(10,6), color= 'blue')
bond['United States'].plot(figsize=(10,6), color = 'darkblue')
bond['China'].plot(figsize=(10,6), color = 'lightblue')
bond['Japan'].plot(figsize=(10,6), color = 'lightblue', ls = '--')

plt.gca().yaxis.set_major_formatter(mtickers.StrMethodFormatter('{x: ,.1f}%'))
plt.legend(loc = 0)
plt.title('10 years Fixed Income rate')
plt.ylabel('Bond Rate')
plt.xlabel('Date')
```

# Daily Cumulative Return analysis

```python
uk['daily_cumulative_return'] = (uk['Price'] / uk.iloc[0]['Price'] - 1) * 100
usa['daily_cumulative_return'] = (usa['Price'] / usa.iloc[0]['Price'] - 1) * 100
china['daily_cumulative_return'] = (china['Price'] / china.iloc[0]['Price'] - 1) * 100
japan['daily_cumulative_return'] = (japan['Price'] / japan.iloc[0]['Price'] - 1) * 100
```

```python
daily_return = pd.concat([uk['daily_cumulative_return'],usa['daily_cumulative_return'],china['daily_cumulative_return'],japan['daily_cumulative_return']], axis = 1)
```

```python
daily_return.columns = ['United Kingdom', 'United States', 'China','Japan']
daily_return = daily_return.dropna()
```

```python
daily_return['United Kingdom'].plot(figsize=(10,6), color= 'blue')
daily_return['United States'].plot(figsize=(10,6), color = 'lightblue')
daily_return['China'].plot(figsize=(10,6), color = 'darkblue')
daily_return['Japan'].plot(figsize=(10,6), color = 'lightblue', ls = '--')
plt.legend()
plt.axhline(0, color='red', linestyle='--')


plt.gca().yaxis.set_major_formatter(mtickers.StrMethodFormatter('{x: ,.1f}%'))
plt.ylabel('Daily Cumulative Return Rate')
plt.title('Daily Cumulative Return of Yield in September ')
```

### Notes: 
##### From the start of month to 12, only China witnessed a rise in yield, this is caused by an expand in GDP growth around 4.8% year on year in Q3, China Briefing stated that "High-tech manufacturing/exports and a widening trade surplus are propping up growth, while domestic demand is soft and the property downturn continues to weigh on spending and investment.". Although property downturn is still a huge weight,  the demand for chinese exports excess the domestic demand, and the economy keep growing leading to higher yield. 
##### For America, Deloitte showed that  "As of August 7, the statutory tariff rate was 18.6%", this will bring the negative impacts for the economy, when other countries also impose tariffs against Trump, leading to imported goods price increase, and those taxations will be passed to consumer leading to lower purchasing power, creating a decline in GDP growth.
##### For UK and Japan, the movement in yield is minor, this signal the economies growth remain stable, so the inflation rate and interest rate just change moderately through out the month.

# US TIPS

```python
tip = pd.read_csv('TIP Historical Data.csv', parse_dates=['Date'])
stip = pd.read_csv('STIP Historical Data.csv', parse_dates=['Date'])
ief = pd.read_csv('IEF Historical Data.csv', parse_dates=['Date'])
```

```python
tip.set_index('Date', inplace= True)
stip.set_index('Date', inplace= True)
ief.set_index('Date', inplace= True)
```

```python
tip['iShares TIPS Bond ETF'] = (tip['Price'] / tip.iloc[0]['Price'] - 1)* 100
stip['iShares 0-5 Year TIPS Bond'] = (stip['Price'] / stip.iloc[0]['Price'] - 1)* 100
```

```python
tip['iShares TIPS Bond ETF'].plot(figsize = (10,6))
stip['iShares 0-5 Year TIPS Bond'].plot(figsize = (10,6), color = 'lightblue')
plt.axhline(0, color='red', linestyle='--', linewidth=1)
plt.legend()
plt.ylabel('Cumulative Rate')
plt.gca().yaxis.set_major_formatter(mtickers.StrMethodFormatter('{x: ,.1f}%'))
```

##### Notes: from 01 to 09 Sep, the TIPS Bond ETF increase significantly, so the inflation is expected to fall, and investors demand more bond to save assets. Between 9 to 17, the TIPS Bond ETF keep stable, so the investors expected higher return from the future increase in inflations. From 17 to the end of September, the TIPS Bond decline toward 0%, this means the real yield increase significantly, and inflation rise, signalling a strong US Economy. The short-term TIPS Bond stay stable between 0% and 0.2% through out the period
